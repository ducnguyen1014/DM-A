```python
NDP_CUSTOMERS = 5
HDP_CUSTOMERS = 10
CARS = 2
FORCE_SAME_CAR = False
```

```python
import numpy as np

np.random.seed(42)


ndp_customers_orders, hdp_customers_orders = [
    [np.random.permutation(customers) for _ in range(100)]
    for customers in [NDP_CUSTOMERS, HDP_CUSTOMERS]
]

ndp_customers_orders, hdp_customers_orders
```

```python
import numpy as np


def segment_customers_orders_by_cars(customers_orders, cars):
    customers_by_cars_solutions = [
        np.array_split(customers_order, cars) for customers_order in customers_orders
    ]
    customers_by_cars_solutions = [
        [list(car) for car in solution] for solution in customers_by_cars_solutions
    ]

    return customers_by_cars_solutions


# Simplified assignment of customer orders by cars
ndp_customers_by_cars, hdp_customers_by_cars = [
    segment_customers_orders_by_cars(customers_orders=orders, cars=CARS)
    for orders in (ndp_customers_orders, hdp_customers_orders)
]
ndp_customers_by_cars
```

```python
def find_edges(solution):
    edges = []
    for car in solution:
        if len(car) > 1:
            for i in range(len(car) - 1):
                edge = [car[i], car[i + 1]]
                edges.append(edge)
    return edges
```

```python
def find_customer(solution, customer):
    for i in range(len(solution)):
        car = solution[i]
        if customer in car:
            return i, car.index(customer)
    raise ValueError("Customer not found.")


print(hdp_solution, find_customer(hdp_solution, 7))
```

```python
def find_customer_after_first_in_hdp_solution(
    hdp_solution, edge, force_same_car=FORCE_SAME_CAR
):
    first_customer_car_index, first_customer_position_index = find_customer(
        solution=hdp_solution, customer=edge[0]
    )
    # Normal: in the same car
    if len(hdp_solution[first_customer_car_index]) > first_customer_position_index + 1:
        return first_customer_car_index, first_customer_position_index + 1
    # If in the next car, get next position as usual
    elif not force_same_car:
        # get the first of next car
        if len(hdp_solution) > first_customer_car_index + 1:
            return first_customer_car_index + 1, 0
        # If in the last car already, get the first car
        else:
            return 0, 0
    else:
        return None, None


print(
    hdp_solution,
    [6, 5],
    find_customer_after_first_in_hdp_solution(hdp_solution, [6, 5], False),
)
```

```python
import copy


def local_search_single_edge(hdp_solution, edge, force_same_car=FORCE_SAME_CAR):
    hdp_solution = copy.deepcopy(hdp_solution)
    first_customer = edge[0]
    second_customer = edge[1]
    customer_after_first_car_index, customer_after_first_customer_index = (
        find_customer_after_first_in_hdp_solution(
            hdp_solution=hdp_solution, edge=edge, force_same_car=force_same_car
        )
    )

    if (
        customer_after_first_car_index is not None
        and customer_after_first_customer_index is not None
    ):
        customer_after_first = hdp_solution[customer_after_first_car_index][
            customer_after_first_customer_index
        ]
        second_customer_car_index, second_customer_position_index = find_customer(
            solution=hdp_solution, customer=second_customer
        )
        hdp_solution[customer_after_first_car_index][
            customer_after_first_customer_index
        ] = second_customer
        hdp_solution[second_customer_car_index][second_customer_position_index] = (
            customer_after_first
        )

    return hdp_solution
```

```python
ndp_solution, hdp_solution = [
    customers_by_cars[0]
    for customers_by_cars in [ndp_customers_by_cars, hdp_customers_by_cars]
]
ndp_solution
```

```python
def local_search(ndp_solution, hdp_solution, force_same_car=FORCE_SAME_CAR):
    edges = find_edges(ndp_solution)
    for edge in edges:
        hdp_solution = local_search_single_edge(
            hdp_solution=hdp_solution, edge=edge, force_same_car=force_same_car
        )
    return hdp_solution


hdp_solution_local_searched = local_search(
    ndp_solution=ndp_solution, hdp_solution=hdp_solution, force_same_car=True
)
print(hdp_solution)
print(ndp_solution)
print(hdp_solution_local_searched)
```
